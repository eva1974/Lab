# Умножение двух амтриц

```python
a = [[9, 19, 44], 
     [31, 9, 5], 
     [78, 7, 99]] 
 
b = [[5, 26, 6, 1], 
     [9, -20, -4, 2], 
     [-1, 3 , 42, 5]] 

m = len(a)   # a: m × n
n = len(b)   # b: n × k
k = len(b[0])

c = [[None for __ in range(k)] for __ in range(m)]    # c: m × k

for i in range(m):
    for j in range(k):       
        c[i][j] = sum(a[i][kk] * b[kk][j] for kk in range(n))
  
print(c)
```
